```python
import pandas as pd
```

```python
hof = 'Resources/basketball_hof_2013_2019.csv'
master = 'Resources/basketball_master.csv'
players_2015 = 'Resources/NBA_Sample_2015.csv'
```

```python
hof_df = pd.read_csv(hof, encoding="ISO-8859-1")
master_df = pd.read_csv(master, encoding="ISO-8859-1")
players_df = pd.read_csv(players_2015, encoding="ISO-8859-1")
```

```python
players_year=players_df[['playerID','year']]
```

```python
players_year['Starting Year'] = players_year.groupby('playerID')['year'].transform('min')
```

```python
players_year['Last Year'] = players_year.groupby('playerID')['year'].transform('max')
```

```python
players_year.to_csv("./Resources/players_year_2015.csv", sep=',', index=False)
```

```python
players_year.drop_duplicates(['playerID'], inplace=True)
```

```python
players_year.drop(['year'],axis=1, inplace=True)
```

```python
players_group = players_df.groupby('playerID').sum().reset_index()
```

```python
players_group.drop(['year'],axis=1, inplace=True)
```

```python
players_group.head()
```

```python
players_group = pd.merge(players_group, players_year, how='left', on=['playerID'])
```

```python
players_group.head()
```

```python
hof_df.rename(columns={'hofID':'playerID'}, inplace=True)
```

```python
master_df.rename(columns={'bioID':'playerID'}, inplace=True)
```

```python
master_df.head()
```

```python
merged_data = pd.merge(master_df, hof_df, how='left', on=['playerID'])
```

```python
merged_data.head()
```

```python
combined_df = pd.merge(players_group, merged_data, how='left', on=['playerID'])
```

```python
combined_df.head()
```

```python
combined_df.rename(columns={'GP':'Total GP'}, inplace=True)
combined_df.rename(columns={'GS':'Total GS'}, inplace=True)
combined_df.rename(columns={'minutes':'Total Minutes'}, inplace=True)
combined_df.rename(columns={'points':'Total Points'}, inplace=True)
combined_df.rename(columns={'oRebounds':'Total Off. Rebounds'}, inplace=True)
combined_df.rename(columns={'dRebounds':'Total Def. Rebound'}, inplace=True)
combined_df.rename(columns={'rebounds':'Total Rebounds'}, inplace=True)
combined_df.rename(columns={'assists':'Total Assists'}, inplace=True)
combined_df.rename(columns={'steals':'Total Steals'}, inplace=True)
combined_df.rename(columns={'blocks':'Total Blocks'}, inplace=True)
combined_df.rename(columns={'turnovers':'Total Turnovers'}, inplace=True)
combined_df.rename(columns={'PF':'Total PF'}, inplace=True)
```

```python
combined_df.head()
```

```python
combined_df.rename(columns={'year':'Year Inducted'}, inplace=True)
```

```python
combined_df['Years Played']=combined_df['Last Year'] - combined_df['Starting Year']+1
```

```python
combined_df['GP per Year']=combined_df['Total GP']/combined_df['Years Played']
```

```python
combined_df['Minutes per Game']=combined_df['Total Minutes'] / combined_df['Total GP']
combined_df['Points per Game']=combined_df['Total Points'] / combined_df['Total GP']
combined_df['O. Rebounds per Game']=combined_df['Total Off. Rebounds'] / combined_df['Total GP']
combined_df['D. Rebounds per Game']=combined_df['Total Def. Rebound'] / combined_df['Total GP']
combined_df['Rebounds per Game']=combined_df['Total Rebounds'] / combined_df['Total GP']
combined_df['Assists per Game']=combined_df['Total Assists'] / combined_df['Total GP']
combined_df['Steals per Game']=combined_df['Total Steals'] / combined_df['Total GP']
combined_df['Blocks per Game']=combined_df['Total Blocks'] / combined_df['Total GP']
combined_df['Turnovers per Game']=combined_df['Total Turnovers'] / combined_df['Total GP']
combined_df['PF per Game']=combined_df['Total PF'] / combined_df['Total GP']
```

```python
combined_df.dtypes
```

```python
combined_df=combined_df.drop(['Total GS','Total Minutes','Total Off. Rebounds','Total Def. Rebound','Total Rebounds','Total Assists','Total Steals','Total Blocks',
                              'Total Turnovers','Total PF','MVP','pos','height','weight','college','birthDate','deathDate','race','name','category','Years Played'], axis=1)
```

```python
combined_df.head()
```

```python
combined_df['Years Retired']=(2015 - combined_df['Last Year'])
combined_df = combined_df.loc[(combined_df['Years Retired']>=4) & (combined_df['Years Retired']<=10)]
combined_df = combined_df.loc[(combined_df['Total GP']>=400)]
combined_df.fillna(0, inplace=True)
combined_df.head()
 
```

```python
combined_df.loc[combined_df['Year Inducted'] > 2015, 'Year Inducted'] = 0
```

```python
combined_df = combined_df.loc[(combined_df['Year Inducted']==0)]
```

```python
combined_df.to_csv("./Resources/BB_clean_2015.csv", sep=',', index=False)
```
