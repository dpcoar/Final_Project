```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
title = "NBA Players - Basic Season Stats (1950-2017).xlsx"
```

```python
data = pd.read_excel(title)
```

```python
data.head()
```

```python
cleaned = data.dropna()
print(cleaned.dtypes)
```

```python
cleaned.head()
```

```python
s = cleaned['Salary in 2019 Dollars']
pd.to_numeric(s, downcast='float')
```

```python
#feature selection
X = cleaned.iloc[:,6:]
y = cleaned['Salary in 2019 Dollars']
columnsDrop = ['Tm', 'TOV','TOV%', 'VORP', 'USG%']
```

```python
# I dont need team names
X.drop(columns = columnsDrop, inplace = True)
```

```python
# ready for the train test split
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression(fit_intercept=True, )
model.fit(X_train, y_train)
```

```python
print(model.intercept_)
print(model.coef_)
print(model.get_params())
```

```python
y_pred = model.predict(X_test)

# Fitting our model with all of our features in X
model.fit(X, y)

score = model.score(X, y)
print(f"R2 Score: {score}")
#this is a bad score
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted':y_pred})
df
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('measured')
ax.set_ylabel('predicted')
plt.show()
```
