```python
import pandas as pd
```

```python
hof = 'Resources/basketball_hof.csv'
master = 'Resources/basketball_master.csv'
players = 'Resources/basketball_players.csv'
```

```python
hof_df = pd.read_csv(hof, encoding="ISO-8859-1")
```

```python
master_df = pd.read_csv(master, encoding="ISO-8859-1")
```

```python
players_df = pd.read_csv(players, encoding="ISO-8859-1")
```

```python
hof_df.head()
```

```python
master_df.head()
```

```python
players_df.head()
```

```python
players_year=players_df[['playerID','year']]
```

```python
players_year['Starting Year'] = players_year.groupby('playerID')['year'].transform('min')
```

```python
players_year['Last Year'] = players_year.groupby('playerID')['year'].transform('max')
```

```python
players_year.to_csv("./Resources/players_year.csv", sep=',', index=False)
```

```python
players_year.drop_duplicates(['playerID'], inplace=True)
```

```python
players_year.drop(['year'],axis=1, inplace=True)
```

```python
players_group = players_df.groupby('playerID').sum().reset_index()
```

```python
players_group.drop(['year'],axis=1, inplace=True)
```

```python
players_group.head()
```

```python
players_group = pd.merge(players_group, players_year, how='left', on=['playerID'])
```

```python
players_group.head()
```

```python
hof_df.rename(columns={'hofID':'playerID'}, inplace=True)
```

```python
master_df.rename(columns={'bioID':'playerID'}, inplace=True)
```

```python
master_df.head()
```

```python
merged_data = pd.merge(master_df, hof_df, how='left', on=['playerID'])
```

```python
merged_data.head()
```

```python
merged_data.dtypes
```

```python
combined_df = pd.merge(players_group, merged_data, how='left', on=['playerID'])
```

```python
combined_df.head()
```

```python
combined_df.drop(['name','category','stint'],axis=1, inplace=True)
```

```python
combined_df['Total GP']=combined_df['GP'] + combined_df['PostGP']
combined_df['Total GS']=combined_df['GS'] + combined_df['PostGS']
combined_df['Total Minutes']=combined_df['minutes'] + combined_df['PostMinutes']
combined_df['Total Points']=combined_df['points'] + combined_df['PostPoints']
combined_df['Total Off. Rebounds']=combined_df['oRebounds'] + combined_df['PostoRebounds']
combined_df['Total Def. Rebounds']=combined_df['dRebounds'] + combined_df['PostdRebounds']
combined_df['Total Rebounds']=combined_df['rebounds'] + combined_df['PostRebounds']
combined_df['Total Assists']=combined_df['assists'] + combined_df['PostAssists']
combined_df['Total Steals']=combined_df['steals'] + combined_df['PostSteals']
combined_df['Total Blocks']=combined_df['blocks'] + combined_df['PostBlocks']
combined_df['Total Turnovers']=combined_df['turnovers'] + combined_df['PostTurnovers']
combined_df['Total PF']=combined_df['PF'] + combined_df['PostPF']
combined_df['Total FG Attempted']=combined_df['fgAttempted'] + combined_df['PostfgAttempted']
combined_df['Total FG Made']=combined_df['fgMade'] + combined_df['PostfgMade']
combined_df['Total FT Attempted']=combined_df['ftAttempted'] + combined_df['PostftAttempted']
combined_df['Total FT Made']=combined_df['ftMade'] + combined_df['PostftMade']
combined_df['Total Three Attempted']=combined_df['threeAttempted'] + combined_df['PostthreeAttempted']
combined_df['Total Three Made']=combined_df['threeMade'] + combined_df['PostthreeMade']
```

```python
combined_df.drop(['GP','PostGP','GS','PostGS','minutes','PostMinutes','points','PostPoints','oRebounds','PostoRebounds','dRebounds','PostdRebounds','rebounds','PostRebounds','assists',
                 'PostAssists','steals','PostSteals','blocks','PostBlocks','turnovers','PostTurnovers','PF','PostPF','fgAttempted','PostfgAttempted','fgMade','PostfgMade','ftAttempted','PostftAttempted',
                  'ftMade','PostftMade','threeMade','PostthreeMade','threeAttempted','PostthreeAttempted'],axis=1, inplace=True)
```

```python
combined_df['Total FG Points']=combined_df['Total FG Made'] * 2
```

```python
combined_df['Total Three Made Points']=combined_df['Total Three Made'] * 3
```

```python
combined_df['Total FG %']=combined_df['Total FG Made']/combined_df['Total FG Attempted']
combined_df['Total FT %']=combined_df['Total FT Made']/combined_df['Total FT Attempted']
combined_df['Total Three %']=combined_df['Total Three Made']/combined_df['Total Three Attempted']
```

```python
combined_df.head()
```

```python
combined_df.rename(columns={'year':'Year Inducted'}, inplace=True)
```

```python
combined_df['Years Played']=combined_df['Last Year'] - combined_df['Starting Year']+1
```

```python
combined_df['GP per Year']=combined_df['Total GP']/combined_df['Years Played']
```

```python
combined_df['Minutes per Game']=combined_df['Total Minutes'] / combined_df['Total GP']
combined_df['Points per Game']=combined_df['Total Points'] / combined_df['Total GP']
combined_df['O. Rebounds per Game']=combined_df['Total Off. Rebounds'] / combined_df['Total GP']
combined_df['D. Rebounds per Game']=combined_df['Total Def. Rebounds'] / combined_df['Total GP']
combined_df['Rebounds per Game']=combined_df['Total Rebounds'] / combined_df['Total GP']
combined_df['Assists per Game']=combined_df['Total Assists'] / combined_df['Total GP']
combined_df['Steals per Game']=combined_df['Total Steals'] / combined_df['Total GP']
combined_df['Blocks per Game']=combined_df['Total Blocks'] / combined_df['Total GP']
combined_df['Turnovers per Game']=combined_df['Total Turnovers'] / combined_df['Total GP']
combined_df['PF per Game']=combined_df['Total PF'] / combined_df['Total GP']
```
